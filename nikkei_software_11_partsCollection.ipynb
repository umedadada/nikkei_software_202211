```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)

# 平均 50, 標準偏差 10 の正規乱数を1,000件生成
x = np.random.normal(50, 10, 5000)

fig,axes = plt.subplots(nrows=2,ncols=3,figsize=(15,15))

axes[0,0].hist(x, bins=10)
axes[0,0].set_title('n=5000,bins=10')

axes[0,1].hist(x, bins=50)
axes[0,1].set_title('n=5000,bins=50')

axes[0,2].hist(x, bins=100)
axes[0,2].set_title('n=5000,bins=100')

# 平均 50, 標準偏差 10 の正規乱数を1.000,000件生成
x = np.random.normal(50, 10, 1000000)

axes[1,0].hist(x, bins=10)
axes[1,0].set_title('n=1000000,bins=10')

axes[1,1].hist(x, bins=50)
axes[1,1].set_title('n=1000000,bins=50')

axes[1,2].hist(x, bins=100)
axes[1,2].set_title('n=1000000,bins=100')
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)

# 平均 50, 標準偏差 10 の正規乱数を1.000,000件生成
x0 = np.random.normal(50, 10, 1000000)
# 平均 100, 標準偏差 10 の正規乱数を1.000,000件生成
x1 = np.random.normal(100, 10, 1000000)
# 平均 50, 標準偏差 5 の正規乱数を1.000,000件生成
x2 = np.random.normal(50, 5, 1000000)
# 平均 50, 標準偏差 30 の正規乱数を1.000,000件生成
x3 = np.random.normal(50, 30, 1000000)

plt.figure()
plt.hist(x0, bins=100, alpha=0.5, histtype='stepfilled', color='red')
plt.hist(x1, bins=100, alpha=0.3, histtype='stepfilled', color='green')
plt.hist(x2, bins=100, alpha=0.3, histtype='stepfilled', color='blue')
plt.hist(x3, bins=100, alpha=0.3, histtype='stepfilled', color='orange')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
np.random.seed(1234)

# 平均 50, 標準偏差 10 の正規乱数を1.000,000件生成
x = np.random.normal(50, 10, 1000000)
centerd_x = x - x.mean()
standard_x = centerd_x / x.std()
 
plt.figure()
plt.hist(x0, bins=100, alpha=0.5, histtype='stepfilled', color='red')
plt.hist(centerd_x, bins=100, alpha=0.3, histtype='stepfilled', color='green')
plt.hist(standard_x, bins=100, alpha=0.3, histtype='stepfilled', color='blue')
plt.show()
```

```python
# t分布の計算
from scipy.stats import t
import numpy as np

# グラフ可視化
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["font.family"]     = "sans-serif"
rcParams["font.sans-serif"] = "Hiragino Maru Gothic Pro"



"""
****************************************
t分布
****************************************
"""
# グラフ設定
fig,ax = plt.subplots(1,1, figsize=(10,8))

# X軸
X = np.linspace(-5,5,100)

# 自由度 n = 1〜5までを用意
n_deg = [14]

# 自由度 n=1〜5までのグラフを描画
for i in n_deg:
    ax.plot(X, t(i).pdf(X), linestyle='-', label="n={}".format(i))


"""
****************************************
グラフ書式設定
****************************************
"""    
plt.xlim(-5, 5)              # X軸範囲
plt.ylim(0, 0.4)             # Y軸範囲
plt.xticks(fontsize= 12)     # X軸フォントサイズ
plt.yticks(fontsize= 12)     # Y軸フォントサイズ
plt.axvline(x=t.ppf(q=0.975,df=14), color='g', linestyle=':', linewidth=3)
plt.axvline(x=t.ppf(q=0.025,df=14), color='g', linestyle=':', linewidth=3)
plt.title("t-distribution", size=16)   # タイトル
plt.legend()                 # 凡例
plt.show()
```
